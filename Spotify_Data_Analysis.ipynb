# Data Cleaning

```python
df1 = df.dropna()
```

```python
df1.isnull().sum()
```

```python
#setting up index should release

df1 = df1.set_index("release_date")
df1.index = pd.to_datetime(df1.index)
df1.head()
```

```python
#Convert duration of miliseconds to seconds

df1["duration"] = df1["duration_ms"].apply(lambda x : round(x/1000))     # As we know in 1 second there is 1000 miliseconds that's why we divide it by 1000
df1.head()
```

```python
df1.drop("duration_ms", axis=1, inplace = True)        
```

```python
df1.head()
```

```python
# top 5 most popular songs in spotify with popularity more than 90
top_5_popular_songs = df[df["popularity"]>90].sort_values("popularity", ascending=False)
top_5_popular_songs[:5]
```

# Data Visualization

```python
# Checking correaltion between variables

corr_df = df1.drop(["key","mode","explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,5))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="magma", linewidths="0.5")
heatmap.set_title("Correlation heatmap Between Variables")
plt.show()
```

```python
# Taking sample from large dataset
sample_df = df1.sample(int(0.004*len(df1)))       #Here, i am taking only 4% of data from my dataset
len(sample_df)
```

```python
# Checking the relationship between energy and loudness

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="r",marker="*").set(title="Loudness vs Energy")
plt.show()
```

```python
# Checking the Relationship between popularity and acousticness

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="y", marker="*").set(title="Popularity vs Acousticness")
plt.show()
```

```python
# getting the years for visulization

df1["dates"] = df1.index.get_level_values("release_date")
df1.dates = pd.to_datetime(df1.dates)
years = df1.dates.dt.year
years
```

```python
# Number of songs per year

sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
total_duration = df1.duration
plt.figure(figsize=(18,7))
sns.barplot(x=years , y=total_duration, errwidth=False).set(title ="Year vs Duration of Songs ")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.set(style="whitegrid")
x = df.groupby("artists")["popularity"].mean().sort_values(ascending=False).head(10)
ax = sns.barplot(x=x.index, y=x)
ax.set_title('Artists with Best Average Popularity')
ax.set_ylabel('Popularity')
ax.set_xlabel('Artists')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.set(style="whitegrid")
x = df.groupby("name")["popularity"].mean().sort_values(ascending=False).head(10)
ax = sns.barplot(x=x.index, y=x)
ax.set_title('Most Popular Tracks')
ax.set_ylabel('Popularity')
ax.set_xlabel('Track Name')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.set(style="whitegrid")
cols = ["popularity","acousticness","instrumentalness","speechiness","danceability","energy"]
sns.pairplot(df[cols], height = 2.5 )
plt.show()
```
